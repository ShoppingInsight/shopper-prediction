```
#OpenCV import 및 원본 영상 출력

import numpy as np
import cv2 
from google.colab.patches import cv2_imshow

img = cv2.imread('lena256_rgb.jpg')
print("원본 영상")
cv2_imshow(img)
```

```
#그레이 영상 출력

im_gray= cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
print("그레이 영상")
cv2_imshow(im_gray)
```

```
# 영상의 덧셈/뺄셈 연산

num=100

im_add =cv2.add(im_gray, num)
print("덧셈/뺄셈 연산")
cv2_imshow(im_add)
```

```
#역변환 영상

im_sub =cv2.subtract(255,im_gray)
print("역변환")
cv2_imshow(im_sub)
```

```
#임계값 th를 기준으로 이진화한 영상

th= 100

th, im_th = cv2.threshold(im_gray, th, 255, cv2.THRESH_BINARY)
print("이진 영상")
cv2_imshow(im_th)
```

```
#영상의 확대 축소(원하는 크기를 입력하세요), 이웃 화소 보간법

#열의 수 입력
r=800

#행의 수 입력
c=500

im_resize1 = cv2.resize(im_gray, dsize=(r, c), interpolation=cv2.INTER_NEAREST)
print("영상 크기 조절(이웃 화소 보간법)")
cv2_imshow(im_resize1)
```

```
#영상의 확대 축소(원하는 크기를 입력하세요), 양선형 보간법

#열의 수 입력
r=800

#행의 수 입력
c=500


im_resize2 = cv2.resize(im_gray, dsize=(r, c), interpolation=cv2.INTER_LINEAR)
print("영상 크기 조절(양선형 보간법)")
cv2_imshow(im_resize2)
```
