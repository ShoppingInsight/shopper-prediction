```
#OpenCV import 및 원본 영상 출력

import numpy as np
import cv2 
from google.colab.patches import cv2_imshow

img = cv2.imread('lena256_rgb_noise.jpg')
print("원본 영상")
cv2_imshow(img)
```

```
# 3*3 스무딩 필터

filter1=np.array([[1/9, 1/9, 1/9], [1/9, 1/9, 1/9],[1/9, 1/9, 1/9]])
im_fil1=cv2.filter2D(img, -1, filter1)
print("3*3 스무딩 필터")
cv2_imshow(im_fil1)
```

```
# 4*4 스무딩 필터

filter2=np.array([[1/16, 1/16, 1/16, 1/16], [1/16, 1/16, 1/16, 1/16], [1/16, 1/16, 1/16, 1/16], [1/16, 1/16, 1/16, 1/16]])
im_fil2=cv2.filter2D(img, -1, filter2)
print("4*4 스무딩 필터")
cv2_imshow(im_fil2)
```

```
# 5*5 스무딩 필터

filter2=np.array([[1/25, 1/25, 1/25, 1/25, 1/25,], [1/25, 1/25, 1/25, 1/25, 1/25,], [1/25, 1/25, 1/25, 1/25, 1/25,], [1/25, 1/25, 1/25, 1/25, 1/25,], [1/25, 1/25, 1/25, 1/25, 1/25,]])
im_fil2=cv2.filter2D(img, -1, filter2)
print("5*5 스무딩 필터")
cv2_imshow(im_fil2)
```
