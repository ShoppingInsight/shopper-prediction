```
#OpenCV import 및 원본 영상 출력

import numpy as np
import cv2 
from google.colab.patches import cv2_imshow

img = cv2.imread('lena256_rgb.jpg')
print("원본 영상")
cv2_imshow(img)
```

```
im_gray= cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
print("그레이 레벨 원영상")
cv2_imshow(im_gray)
```

```
# 3*3 스무딩 필터

filter1=np.array([[1/9, 1/9, 1/9], [1/9, 1/9, 1/9],[1/9, 1/9, 1/9]])
im_fil1=cv2.filter2D(im_gray, -1, filter1)
print("3*3 스무딩 필터")
cv2_imshow(im_fil1)
```

```
# 5*5 스무딩 필터(3*3과의 차이 확인)

filter2=np.array([[1/25, 1/25, 1/25, 1/25, 1/25,], [1/25, 1/25, 1/25, 1/25, 1/25,], [1/25, 1/25, 1/25, 1/25, 1/25,], [1/25, 1/25, 1/25, 1/25, 1/25,], [1/25, 1/25, 1/25, 1/25, 1/25,]])
im_fil2=cv2.filter2D(im_gray, -1, filter2)
print("5*5 스무딩 필터")
cv2_imshow(im_fil2)
```

```
# 소벨 에지 검출

filter1=np.array([[1, 0, -1], [2, 0, -2],[1, 0, -1]])
im_fil1=cv2.filter2D(im_gray, -1, filter1)
print("수직방향 에지검출(소벨)")
cv2_imshow(im_fil1)

filter1=np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])
im_fil2=cv2.filter2D(im_gray, -1, filter1)
print("수평방향 에지검출(소벨)")
cv2_imshow(im_fil2)

im_fil3 =im_fil1+ im_fil2
print("최종 소벨 에지검출 결과 영상")
cv2_imshow(im_fil3)
```
