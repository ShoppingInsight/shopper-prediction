```python
# 예제

from sklearn.neighbors import KNeighborsClassifier
import numpy as np

# 특징 및 종속 변수 데이터 생성
x=np.array([[3,3.5], [4, 5], [2,4.5],[1,1], [3.5,7], [6,5.5]])
y=np.array([1,1,1,2,2,2])

# kNN 모델 생성
model = KNeighborsClassifier(n_neighbors=3)
model.fit(x, y)    # 학습

# 예측
y_pred = model.predict([[3,2]])
y_pred_proba =  model.predict_proba([[3,2]])
print(y_pred)
print(y_pred_proba)


```

```python
# 예제 2

from sklearn.neighbors import KNeighborsClassifier
import numpy as np

# 특징 및 종속 변수 데이터 생성
x=np.array([[1,6], [2, 5], [2,7],[3,7], [4,1], [4,4], [4,6], [5,2],[6,1],[6,3],[7,1],[7,2],[7,4]])
y=np.array([1,1,1,1,2,1,1,2,2,2,2,2,2])

# kNN 모델 생성
model = KNeighborsClassifier(n_neighbors=5)
model.fit(x, y)    # 학습

# 예측
y_pred = model.predict([[5,4]])
y_pred_proba =  model.predict_proba([[5,4]])
print(y_pred)
print(y_pred_proba)
```

```python
# 예제 3

from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Input data: [height, weight, age]
X = np.array([
    [180, 75, 25],
    [160, 60, 30],
    [170, 68, 28],
    [165, 55, 45],
    [175, 80, 22],
    [158, 52, 50],
    [182, 85, 26],
    [168, 63, 32],
    [172, 78, 24],
    [166, 58, 40]
])

# Body type labels: A, B, C
y = np.array([
    'A', 'B', 'A', 'C', 'A',
    'C', 'A', 'B', 'A', 'B'
])

# Create and train model
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y)

# Predict a new person
predicted_class = knn.predict([[172, 70, 27]])[0]
print("Predicted class:", predicted_class)

# Visualization (3D)
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Color mapping
color_map = {'A': 'red', 'B': 'blue', 'C': 'green'}
point_colors = [color_map[label] for label in y]

print(point_colors)

# Plot training data
ax.scatter(X[:, 0], X[:, 1], X[:, 2],
           c=point_colors, s=60, label='Training Data')

# Plot new data
ax.scatter(172, 70, 27,
           color='gold', marker='*', s=200, label='New Person')

# Labels and title
ax.set_xlabel('Height (cm)')
ax.set_ylabel('Weight (kg)')
ax.set_zlabel('Age')
ax.set_title("3D KNN Classification Result: Predicted Body Type")
ax.legend()
plt.show()
```

```python
# 예제 4

from sklearn.neighbors import KNeighborsRegressor
import numpy as np
import matplotlib.pyplot as plt

# Input data
# Study time (x1), Sleep time (x2)
X = np.array([
    [1, 5],
    [2, 4],
    [3, 3],
    [4, 2],
    [5, 1]
])

# Exam scores (y)
y = np.array([60, 65, 70, 75, 80])

# KNN regression model (k=2)
knn = KNeighborsRegressor(n_neighbors=2)

# Train model
knn.fit(X, y)

# New data: study 3 hours, sleep 4 hours
new_data = np.array([[3, 4]])

# Prediction
result = knn.predict(new_data)
print("Predicted exam score:", result)

# Visualization
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Training data
ax.scatter(X[:, 0], X[:, 1], y, color='blue', s=60, label='Training Data')

# Prediction point
ax.scatter(new_data[:, 0], new_data[:, 1], result,
           color='red', marker='*', s=200, label='Prediction')

# Labels and title
ax.set_xlabel('Study Time')
ax.set_ylabel('Sleep Time')
ax.set_zlabel('Exam Score')
ax.set_title(f'Predicted Exam Score: {result[0]:.2f}')
ax.legend()

plt.show()
```
