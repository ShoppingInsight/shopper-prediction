```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
from konlpy.tag import Okt

# 1. 훈련 데이터
train_texts = [
    '무료 당첨 지금 클릭하세요',
    '당신은 선택되었습니다 지금 확인하세요',
    '안녕하세요 회의 일정 확인 바랍니다',
    '이번 주 회식은 금요일입니다',
    '무료 쿠폰 제공 클릭 필수',
    '지금 가입하면 선물 증정',
    '안건 회의록 첨부합니다',
    '회의 일정 공유드립니다',
    '할인 당첨 클릭 이벤트 참여',
    '업무 관련 문서 확인 요청드립니다'
]
train_labels = ['spam', 'spam', 'ham', 'ham', 'spam', 'spam', 'ham', 'ham', 'spam', 'ham']

# 2. 별도 테스트 데이터 생성
test_texts = [
    '지금 무료 쿠폰을 받으세요',       # 스팸으로 예측되어야 함
    '오늘 회의 자료 첨부드립니다',     # 정상으로 예측되어야 함
    '클릭해서 상품을 확인하세요',     # 스팸
    '업무 보고서 작성 부탁드립니다'    # 정상
]

# 3. 형태소 분석
okt = Okt()
def tokenize(text):
    return ' '.join(okt.morphs(text))

train_tokenized = [tokenize(t) for t in train_texts]
test_tokenized = [tokenize(t) for t in test_texts]

# 4. TF-IDF 벡터화 (학습 데이터로 fit)
vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(train_tokenized)
X_test = vectorizer.transform(test_tokenized)

# 5. 모델 학습
model = MultinomialNB()
model.fit(X_train, train_labels)

# 6. 테스트 문장 예측
y_pred = model.predict(X_test)

# 7. 결과 출력
for i, sentence in enumerate(test_texts):
    print(f"'{sentence}' → 예측 결과: {y_pred[i]}")
```
